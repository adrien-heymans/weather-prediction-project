# Metrics Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the CSV file
file_path = '/content/drive/MyDrive/CSI6900/Final Execution/weather_prediction_metrics.csv'
df = pd.read_csv(file_path)

# Normalize MSE and MAE by feature
features = df['Feature'].unique()
df_normalized = df.copy()
for feature in features:
    df_normalized.loc[df['Feature'] == feature, 'MSE'] = df[df['Feature'] == feature]['MSE'] / df[df['Feature'] == feature]['MSE'].max()
    df_normalized.loc[df['Feature'] == feature, 'MAE'] = df[df['Feature'] == feature]['MAE'] / df[df['Feature'] == feature]['MAE'].max()

# Remove outliers
df_normalized = df_normalized[df_normalized['MSE'] < 1.5]
df_normalized = df_normalized[df_normalized['MAE'] < 1.5]

# Rename features for better visualization
df_normalized['Feature'] = df_normalized['Feature'].str.replace('_', ' ').str.title()

# Replace "CNN" with "1D-CNN" and "Randmoized" with "Randomized training data"
df_normalized['Model'] = df_normalized['Model'].replace({'CNN': '1D-CNN'})
df_normalized['Randomized'] = df_normalized['Randomized'].replace({'Randmoized': 'Randomized training data'})

# Function to remove outliers based on the IQR method
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        df = df[~((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR)))]
    return df

# Remove outliers for MSE, MAE, and R2
df_no_outliers = remove_outliers(df_normalized, ['MSE', 'MAE', 'R2'])

# Filter the dataframe to only include rows where n_future is 365
df_filtered = df_no_outliers

# Function to save plot
def save_plot(fig, name, save_path):
    fig.savefig(f"{save_path}/{name}.png", bbox_inches='tight')
    plt.show()
    plt.close(fig)

# Function to print summary statistics
def print_summary(data, group_by, metrics):
    summary = data.groupby(group_by)[metrics].describe()
    print(summary)

# Plotting functions
def plot_boxplot(data, x, y, hue, title, filename, save_path):
    fig, ax = plt.subplots(figsize=(12, 6))
    sns.boxplot(data=data, x=x, y=y, hue=hue, palette="husl", ax=ax)
    ax.set_title(title)
    save_plot(fig, filename, save_path)
    print_summary(data, [x, hue], [y])

def plot_lineplot(data, x, y, hue, title, filename, save_path):
    fig, ax = plt.subplots(figsize=(12, 6))
    sns.lineplot(data=data, x=x, y=y, hue=hue, palette="husl", ax=ax)
    ax.set_title(title)
    save_plot(fig, filename, save_path)
    print_summary(data, [x, hue], [y])

def plot_heatmap(data, x, y, value, title, filename, save_path):
    fig, ax = plt.subplots(figsize=(12, 6))
    pivot_table = data.pivot_table(values=value, index=y, columns=x, aggfunc='mean')
    sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap="coolwarm", ax=ax)
    ax.set_title(title)
    save_plot(fig, filename, save_path)
    print(pivot_table)

# Function to create the desired adjusted plot
def plot_boxplot_adjusted(data, x, y, hue, title, filename, save_path):
    fig, ax = plt.subplots(figsize=(12, 6))
    sns.boxplot(data=data, x=x, y=y, hue=hue, palette="husl", ax=ax)
    ax.set_title(title)
    save_plot(fig, filename, save_path)
    print_summary(data, [x, hue], [y])

# Create plots folder if not exists
save_path = '/content/drive/MyDrive/CSI6900/Final Execution/plots/metrics'
os.makedirs(save_path, exist_ok=True)

# Plotting the heatmap for number of past and future days
plot_heatmap(df_normalized, 'n_past', 'n_future', 'R2', 'Effect of Previous Days on Future Days (R2)', 'heatmap_past_future_r2', save_path)

# Questions with only n_future = 365:
# 1. Did the randomization of training data help the models to learn?
plot_boxplot(df_filtered, 'Randomized', 'R2', 'Model', 'Effect of Randomized training data on Model Performance (R2)', 'randomization_r2_boxplot', save_path)

# 2. Are LSTM better than 1D-CNNs, or is it the opposite?
plot_boxplot(df_filtered, 'Model', 'R2', 'Feature', 'Performance Comparison: LSTM vs 1D-CNN (R2)', 'lstm_vs_cnn_r2_boxplot', save_path)

# 3. MSE comparison with randomization
plot_boxplot(df_filtered, 'Randomized', 'MSE', 'Model', 'Effect of Randomized training data on Model Performance (MSE)', 'randomization_mse_boxplot', save_path)

# 4. MAE comparison with randomization
plot_boxplot(df_filtered, 'Randomized', 'MAE', 'Model', 'Effect of Randomized training data on Model Performance (MAE)', 'randomization_mae_boxplot', save_path)

# 5. Performance comparison (MSE) of LSTM vs 1D-CNN
plot_boxplot(df_filtered, 'Model', 'MSE', 'Feature', 'Performance Comparison: LSTM vs 1D-CNN (MSE)', 'lstm_vs_cnn_mse_boxplot', save_path)

# 6. Performance comparison (MAE) of LSTM vs 1D-CNN
plot_boxplot(df_filtered, 'Model', 'MAE', 'Feature', 'Performance Comparison: LSTM vs 1D-CNN (MAE)', 'lstm_vs_cnn_mae_boxplot', save_path)

# 7. Performance metrics over time for a specific feature (e.g., Solar Radiation)
fig, axs = plt.subplots(2, 2, figsize=(14, 12))

features = ['Solar Radiation', 'Avg Hourly Pressure Station', 'Avg Hourly Temperature', 'Precipitation']
titles = ['Performance Over Time: Solar Radiation (R2)', 'Performance Over Time: Avg Hourly Pressure Station (R2)',
          'Performance Over Time: Avg Hourly Temperature (R2)', 'Performance Over Time: Precipitation (R2)']

for ax, feature, title in zip(axs.flatten(), features, titles):
    sns.lineplot(data=df_filtered[df_filtered['Feature'] == feature], x='n_past', y='R2', hue='Model', palette="husl", ax=ax)
    ax.set_title(title)

plt.tight_layout()
save_plot(fig, 'combined_lineplots_r2', save_path)

# 8. Performance comparison by type of value predicted and randomization
plot_boxplot(df_filtered, 'Feature', 'R2', 'Randomized', 'Performance Comparison by Feature and Randomized training data (R2)', 'feature_randomization_r2_boxplot', save_path)

# Adjusted plots for better comparison
plot_boxplot_adjusted(df_filtered, 'Model', 'MSE', 'Randomized', 'Effect of Model Type and Randomized training data on MSE', 'adjusted_mse_boxplot_no_outliers', save_path)
plot_boxplot_adjusted(df_filtered, 'Model', 'R2', 'Randomized', 'Effect of Model Type and Randomized training data on R2', 'adjusted_r2_boxplot_no_outliers', save_path)
plot_boxplot_adjusted(df_filtered, 'Model', 'MAE', 'Randomized', 'Effect of Model Type and Randomized training data on MAE', 'adjusted_mae_boxplot_no_outliers', save_path)
```

# Predictions Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
from dateutil import parser
import os
import pandas as pd

# General plotting function for any column pair
def plot_comparison(data, model, n_past, n_future, actual_col, prediction_col, title, ax):
    subset = data[(data['Model'] == model) & (data['n_past'] == n_past) & (data['n_future'] == n_future)]
    sns.lineplot(data=subset, x='Date', y=actual_col, label='Actual', ax=ax)
    sns.lineplot(data=subset, x='Date', y=prediction_col, label=f'{model} Prediction', ax=ax)
    ax.set_title(title)
    ax.legend()

# Function to plot actual values vs predictions for a specified model for all relevant columns
def plot_predictions(data, model, n_past, n_future, save_path):
    subset = data[(data['Model'] == model) & (data['n_past'] == n_past) & (data['n_future'] == n_future)]

    plt.figure(figsize=(14, 14))

    columns = [
        ('Actual_avg_hourly_temperature', 'avg_hourly_temperature', 'Average Hourly Temperature'),
        ('Actual_avg_hourly_pressure_station', 'avg_hourly_pressure_station', 'Average Hourly Pressure Station'),
        ('Actual_solar_radiation', 'solar_radiation', 'Solar Radiation'),
        ('Actual_precipitation', 'precipitation', 'Precipitation')
    ]

    for i, (actual_col, prediction_col, title) in enumerate(columns, 1):
        ax = plt.subplot(2, 2, i)
        plot_comparison(data, model, n_past, n_future, actual_col, prediction_col, title, ax)

    plt.tight_layout()
    plt.savefig(f"{save_path}/{model}_predictions.png")
    plt.show()

# Function to compare different models for all relevant columns
def compare_models(data, n_past, n_future, save_path):
    models = data['Model'].unique()
    plt.figure(figsize=(14, 14))

    columns = [
        ('Actual_avg_hourly_temperature', 'avg_hourly_temperature', 'Average Hourly Temperature'),
        ('Actual_avg_hourly_pressure_station', 'avg_hourly_pressure_station', 'Average Hourly Pressure Station'),
        ('Actual_solar_radiation', 'solar_radiation', 'Solar Radiation'),
        ('Actual_precipitation', 'precipitation', 'Precipitation')
    ]

    for i, (actual_col, prediction_col, title) in enumerate(columns, 1):
        ax = plt.subplot(2, 2, i)
        for model in models:
            subset = data[(data['Model'] == model) & (data['n_past'] == n_past) & (data['n_future'] == n_future)]
            sns.lineplot(data=subset, x='Date', y=prediction_col, label=f'{model} Prediction', ax=ax)
        actual = data[(data['Model'] == models[0]) & (data['n_past'] == n_past) & (data['n_future'] == n_future)]
        sns.lineplot(data=actual, x='Date', y=actual_col, label='Actual', ax=ax)
        ax.set_title(title)
        ax.legend()

    plt.tight_layout()
    plt.savefig(f"{save_path}/model_comparison.png")
    plt.show()

# Function to compare randomized vs non-randomized data for a model for all relevant columns
def compare_randomization(data, model, n_past, n_future, save_path):
    subset = data[(data['Model'] == model) & (data['n_past'] == n_past) & (data['n_future'] == n_future)]
    randomized = subset[subset['Randomized'] == True]
    non_randomized = subset[subset['Randomized'] == False]

    plt.figure(figsize=(14, 14))

    columns = [
        ('Actual_avg_hourly_temperature', 'avg_hourly_temperature', 'Average Hourly Temperature'),
        ('Actual_avg_hourly_pressure_station', 'avg_hourly_pressure_station', 'Average Hourly Pressure Station'),
        ('Actual_solar_radiation', 'solar_radiation', 'Solar Radiation'),
        ('Actual_precipitation', 'precipitation', 'Precipitation')
    ]

    for i, (actual_col, prediction_col, title) in enumerate(columns, 1):
        ax = plt.subplot(2, 2, i)
        sns.lineplot(data=subset, x='Date', y=actual_col, label='Actual', ax=ax)
        sns.lineplot(data=randomized, x='Date', y=prediction_col, label=f'{model} Prediction (Randomized Training Data)', ax=ax)
        sns.lineplot(data=non_randomized, x='Date', y=prediction_col, label=f'{model} Prediction (Non-Randomized Training Data)', ax=ax)
        ax.set_title(title)
        ax.legend()

    plt.tight_layout()
    plt.savefig(f"{save_path}/{model}_randomization_comparison.png")
    plt.show()

# Function to compare randomized vs non-randomized data overall for all relevant columns
def compare_randomization_overall(data, n_past, n_future, save_path):
    subset = data[(data['n_past'] == n_past) & (data['n_future'] == n_future) & (data['Model']!= 'Baseline')]
    randomized = subset[subset['Randomized'] == True]
    non_randomized = subset[subset['Randomized'] == False]

    plt.figure(figsize=(14, 14))

    columns = [
        ('Actual_avg_hourly_temperature', 'avg_hourly_temperature', 'Average Hourly Temperature'),
        ('Actual_avg_hourly_pressure_station', 'avg_hourly_pressure_station', 'Average Hourly Pressure Station'),
        ('Actual_solar_radiation', 'solar_radiation', 'Solar Radiation'),
        ('Actual_precipitation', 'precipitation', 'Precipitation')
    ]

    for i, (actual_col, prediction_col, title) in enumerate(columns, 1):
        ax = plt.subplot(2, 2, i)
        sns.lineplot(data=subset, x='Date', y=actual_col, label='Actual', ax=ax)
        sns.lineplot(data=randomized, x='Date', y=prediction_col, label='Randomized Training Data', ax=ax)
        sns.lineplot(data=non_randomized, x='Date', y=prediction_col, label='Non-Randomized Training Data', ax=ax)
        ax.set_title(title)
        ax.legend()

    plt.tight_layout()
    plt.savefig(f"{save_path}/randomization_overall.png")
    plt.show()

# Create plots folder if not exists
save_path = '/content/drive/MyDrive/CSI6900/Final Execution/plots/predictions'
os.makedirs(save_path, exist_ok=True)

# Importing the data
data = pd.read_csv('/content/drive/MyDrive/CSI6900/Final Execution/weather_predictions.csv')
data['Date'] = data['Date'].apply(parser.parse)
data = data.sort_values('Date')

# Example usage
n_past = 90
n_future = 365

# Overall randomization comparison
compare_randomization_overall(data, n_past, n_future, save_path)

# Compare models
compare_models(data, n_past, n_future, save_path)

# Compare randomization for CNN model
compare_randomization(data, 'CNN', n_past, n_future, save_path)
compare_randomization(data, 'LSTM', n_past, n_future, save_path)

# Plot specific model predictions
plot_predictions(data, 'CNN', n_past, n_future, save_path)
plot_predictions(data, 'LSTM', n_past, n_future, save_path)
```
